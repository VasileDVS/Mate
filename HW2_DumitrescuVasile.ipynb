## Homework for topic 1.2. Calculating costs using matrices

**Task 1**

```python
import numpy as np
A = np.array(
    [[-1, 1, 2],
     [0, -1, -3],
     [4, -3, 2],
     ]
    )
B = np.array(
    [[1],
     [-4],
     [7],
     ]
    )
print(f"Determinant of matrix A: {round(np.linalg.det(A))}")
A_back = np.linalg.inv(A)

print(f"Inverse matrix of matrix A: \r\n {A_back }")

print(f"Solution column vector: \r\n {A_back.dot(B)}")
```

```python
import numpy as np
A = np.array(
    [[3, -1, 0],
     [2, -2, 4],
     [5, -6, 2],
     ]
    )
B = np.array(
    [[4],
     [8],
     [13],
     ]
    )
print(f"Determinant of matrix A: {round(np.linalg.det(A))}")
A_back = np.linalg.inv(A)

print(f"Inverse matrix of matrix A: \r\n {A_back }")

print(f"Solution column vector: \r\n {A_back.dot(B)}")
```
