## Use all available parameters for forecasting


```python
real_estate_data.columns
x_names =['X1 transaction date','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude']
X = real_estate_data[x_names]
X.head()
```

## Define the forecasting objective (price)

```python
y = real_estate_data['Y house price of unit area']
y.head()
```

## Write a model on the DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

# Define the module. Specify the number for random_state to guarantee the same results every time you run the code
real_estate_model = DecisionTreeRegressor(random_state=1)

# Fit the model!
real_estate_model.fit(X, y)
```

## Make a price forecast for the first 5 rows

```python
print("We make a forecast for the first 5 houses:")
# print(X.head())
print("The following are the prices that we can forecast:")
print(real_estate_model.predict(X.head()))
print(y.head())
```
