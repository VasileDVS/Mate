3. Goods in the industry are produced at a speed that can be described by the vector (1,5,2,6,7). Due to negative force majeure events, changes for the current year can be characterized as (0,-1,0,-0,-1), but the industry still maintains a growth of 20% per year. How long will it take for the industry to completely overcome the consequences of the force majeure circumstances?

````python
import numpy as np
# Define vectors
a = np.array([1, 5, 2, 6, 7])

b = np.array([0, -1, 0, -0, -1])

# Check if all elements of the vector are positive
def pos(A):
  flag = 1
  for el in A:
    if el<=0:
      flag = 0
  return flag
# Determine the rate according to the task
da = a*0.2
print(da)

change = np.array([0,0,0,0,0])
mb = -b
i = 0
# Until the difference vector becomes positive
while not np.all(change-mb >= 0):
  # Add changes corresponding to the growth rate
  change =change + da
  print(change-mb)
  i+=1

# As a result, we get how many iterations it takes for the difference vector to become positive
print(f"The crisis will be overcome completely for all goods in the industry in {i} years")
````
