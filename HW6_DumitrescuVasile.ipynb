Calculate the derivative of the employee work efficiency function.

```python
from sympy import diff, symbols
x, y = symbols('x y')

print(diff(-3*x*x+30*x))
```

Draw the graph of the function and the graph of the derivative on one plot.

```python
import matplotlib.pyplot as plt
from sympy import diff, symbols

fig, ax = plt.subplots()

# Move the left and bottom spines to x = 0 and y = 0 respectively
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Hide the top and right spine
ax.spines[["top", "right"]].set_visible(False)

# Draw arrows (as black triangles: ">k"/"^k") at the ends of the axes
# Also, turn off clipping for the arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Add grid lines with a dashed style.
ax.grid(True, linestyle='-.')

# Generate a series of x values. 100 elements from -2 to 2

x, y = symbols('x y')
# Define the function
fx = -3*x*x+30*x
# Calculate the derivative of the function
dx = diff((fx))
k = []
n = []
l = []

for i in range(800):
    j = (i)*0.01
    k.append(j)
    n.append(dx.subs(x,j))
    l.append(fx.subs(x,j))


ax.plot(k,n)
ax.plot(k,l)


# Show the plot
plt.show()
```

Calculate the value of the maximum function analytically.

$y = -3x^2+30x$  
$y' = -6x+30$  

$y' = 0$ (extrema condition)  
$-6x+30=0$   
$x= \dfrac{30}{6} = 5$  
the function reaches its maximum at this point: $x=5$  
The value of the function at this point:  
$y = -3 \cdot 5^2+30\cdot5=75$

Draw the graph of the work efficiency function, marking the point of maximum efficiency.

```python
import numpy as np
from scipy.optimize import minimize_scalar, fminbound
def f(x): # we'll draw arrows (as black triangles: ">k"/"^k") at the ends of the axes
    return -3*(x)**2+30*x
res = minimize_scalar(f, bounds=(0,8), method='bounded') # find the minimum of f(x) within the specified bounds. Search criterion is minimum, feasible set of x from 0 to 8.
print ("argmin=",res.x)

argmin = fminbound(f, 0, 8) # or this way
print ("argmin=", argmin)
argmax = fminbound(lambda x: -f(x), 0,8) # find the maximum of f(x) within the specified bounds
print ("argmax=", argmax)

import matplotlib.pyplot as plt
x=np.linspace(0,8,100)
plt.plot(x,f(x),'k') # plot
plt.scatter([argmin,argmax], [f(argmin),f(argmax)], linewidths=[3,3]) # local extrema
plt.xlabel('x');plt.ylabel('y');plt.grid();plt.show()
```

```python
import numpy as np
from scipy.integrate import quad
f = lambda x, a: -3*(x)**2+30*x # function
print(quad(f, 0, 8, args=(2,))) # integration result and estimation of the absolute error of the result
```
